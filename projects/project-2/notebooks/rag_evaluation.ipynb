```python
queries = [
    "investimento agressivo sem risco",
    "cliente conservador investindo em ações",
    "produto com garantia de retorno"
]

for q in queries:
    print("\n==============================")
    print("QUERY:", q)

    docs = retrieve(q)

    print("\nANTES (retrieve):")
    for d in docs:
        print("-", d["text"])

    docs_reranked = rerank(q, docs)

    print("\nDEPOIS (rerank):")
    for d in docs_reranked:
        print("-", d["text"])
```

## Análise de Re-ranking

O reranker melhora a ordenação dos documentos retornados pelo retrieval,
priorizando os mais relevantes com base na consulta do usuário.

Antes do re-ranking, os documentos recuperados nem sempre estão na melhor ordem.
Após o re-ranking, os trechos mais relevantes aparecem nas primeiras posições,
melhorando a qualidade da resposta fornecida pelo sistema RAG.
